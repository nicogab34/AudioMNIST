```python
import matplotlib.pyplot as plt
import glob
import os

filenames = glob.glob("../logs/*.dat")
latest = max(filenames, key=os.path.getctime)
with open(latest) as mem_fh:
    next(mem_fh)
    time = []
    mem_usage = []
    for l in mem_fh:
        values = l.split()
        mem_usage.append(float(values[1]))
        time.append(float(values[2]))
```

```python
plt.plot(time, mem_usage)
```
